## Inventario de datasets disponibles

```python
import os
from pathlib import Path

carpeta_datos = Path('../datos')

archivos = []
for f in sorted(carpeta_datos.iterdir()):
    if f.is_file() and f.suffix in ['.csv', '.xlsx', '.xls']:
        size_kb = f.stat().st_size / 1024
        archivos.append({'archivo': f.name, 'extension': f.suffix, 'tamaño_kb': round(size_kb, 1)})

df_inventario = pd.DataFrame(archivos)
print(f"Total de archivos de datos: {len(df_inventario)}")
df_inventario
```
