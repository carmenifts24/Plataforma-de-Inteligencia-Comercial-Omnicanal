```python
# Importar las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar tu dataset
# Reemplazá por tu propio archivo si lo necesitás

df = pd.read_csv("../datos/01_raw/Sample - Superstore.csv", encoding="latin1")



# -----------------------
# 1. Información general
# -----------------------

print("▶️ Primeras filas:")
print(df.head())  # Muestra las 5 primeras filas

print("\n▶️ Últimas filas:")
print(df.tail())  # Muestra las 5 últimas filas

print("\n▶️ Dimensiones del dataset:")
print(df.shape)  # (filas, columnas)

print("\n▶️ Nombres y tipos de columnas:")
print(df.dtypes)

```

```python
# -----------------------
# 2. Nulos y duplicados
# -----------------------

print("\n▶️ Cantidad de valores nulos por columna:")
print(df.isnull().sum())

print("\n▶️ Porcentaje de nulos por columna:")
print((df.isnull().mean() * 100).round(2))

print("\n▶️ Filas duplicadas:")
print(df.duplicated().sum())

```

```python
# -----------------------
# 3. Estadísticas descriptivas
# -----------------------

print("\n▶️ Estadísticas básicas:")
print(df.describe())  # Solo numéricas

```

```python
# -----------------------
# 4. Variables categóricas
# -----------------------

print("\n▶️ Columnas categóricas:")
cat_cols = df.select_dtypes(include='object').columns
print(cat_cols)

for col in cat_cols:
    print(f"\n📌 Frecuencia en '{col}':")
    print(df[col].value_counts())


```

```python
# -----------------------
# 5. Outliers y distribución
# -----------------------

# Histograma para cada variable numérica
df.select_dtypes(include=np.number).hist(figsize=(12, 8), bins=20)
plt.suptitle("Histograma de variables numéricas")
plt.tight_layout()
plt.show()

# Boxplot por variable
for col in df.select_dtypes(include=np.number).columns:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot: {col}")
    plt.show()

```

```python
# -----------------------
# 6. Correlación
# -----------------------

corr = df.select_dtypes(include=np.number).corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlación")
plt.show()
```
